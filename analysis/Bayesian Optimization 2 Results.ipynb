```python
EXP_DIR = "../experiments/bayes_opt_2/"
```

```python
import pandas as pd
import json
from glob import glob
import os

example={"accuracy": 0.6869158878504673,
    "confusion_matrix": "[[  6  20   2   6]\n [  0  27   4   5]\n [  1   4   6  12]\n [  5   7   1 108]]",
    "precisionmicro": 0.6869158878504673,
    "precisionmacro": 0.5628707959584506,
    "precisionweighted": 0.673503413532147,
    "recallmicro": 0.6869158878504673,
    "recallmacro": 0.51997553423094,
    "recallweighted": 0.6869158878504673,
    "f1micro": 0.6869158878504673,
    "f1macro": 0.5064534601999912,
    "f1weighted": 0.6585577974867919}

df = pd.DataFrame(columns=['dataset', 'model', 'optimizer'] + list(example.keys()))


for dataset in ['4_class_tt', 'die_vs_all_tt']:
    for model_dir in glob(EXP_DIR + dataset + "/*"):
        for optimizer in ['Adam', 'SGD']:
            path = f"{model_dir}/{optimizer}/BEST/testperformance_metrics.json"
            try:
                with open(path,'r') as f:
                    metrics = json.load(f)
                    metrics['dataset'] = dataset
                    metrics['model'] = os.path.basename(model_dir)
                    metrics['optimizer'] = optimizer
#                     print(metrics)
                    print
                    df = df.append(metrics, ignore_index=True)
            except:
                pass
#                 print("Skipping", path)

        
df = df.drop(['confusion_matrix'], axis=1).sort_values(['accuracy'], ascending=False)
```
