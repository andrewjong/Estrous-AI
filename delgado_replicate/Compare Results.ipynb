```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
from glob import glob


EXP_DIR = "../experiments/delgado/"
results = glob(EXP_DIR + "**/*.json")
```

```python
import json


summaries = []

for r in results:
    with open(r, 'r') as f:
        d = json.load(f)
        summaries.append(d)
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)

df=pd.DataFrame(summaries)
df = df.set_index(["class_groups", "model", "dataset", ]).sort_index()
df = df[["best_test", "best_train", "best_validation", "duration", "best_parameters"]]
df
```

```python
df.to_csv("experimental_results.csv")
```

```python
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: lightgreen' if v else '' for v in is_max]
```

```python
for i in df.index.levels[0]:
    df.loc[i][["best_test"]].style.apply(highlight_max)
```

# Observations and Comments


GAN Augmentation did not give a clear accuracy boost over the base training set. Would not recommend because actually increases training and optimization time.

Best test accuracy is consistently 3-4 percentage points higher than the best validation found during hyperparameter optimization. This is likely because during the final evaluation on the test set, the model has access to the full training set, not a validation held-out subset.

Diestrus v. all does 2-3 percentage points better than met_die v pro_est. Guessing because the dataset is better balanced in Die v. all.
